#openCVを使い画像から顔検出しクロップし保存する
https://qiita.com/FrozenVoice/items/d1dc5f3d780f5097bfab

```
import numpy as np
import cv2
import io,os
#顔を検出したいディレクトリ名
input_dir="./downloads/テスト"
#顔を切り出した画像の保存先ディレクトリ名
crop_dir="./crop_img"
#元画像に検出範囲を書き込んだ画像の保存先ディレクトリ名
surrounded_dir="./surrounded_img"
#ディレクトリに格納されているファイル一覧取得
file_list = os.listdir('./downloads/テスト')
#cv2で使う重みファイルがあるディレクトリ
#重みファイルはgithubからcloneでなくpipなどでインストールしたほうが確実
cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt2.xml")
```

```
#顔が一つも検出されなかった画像の数カウンタ
n=0
#顔検出されなかった画像のリスト
n_files=[]
for file_name in file_list:
    root, ext = os.path.splitext(file_name)
    if ext == u'.png' or u'.jpeg' or u'.jpg':
      abs_name = input_dir  + '/'+file_name
      #画像ファイルパスから画像の読み込み
      img = cv2.imread(abs_name,cv2.IMREAD_COLOR)
      #画像のグレースケール化
      #img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      #検出された顔らの情報がfaceに入る
      face=cascade.detectMultiScale(img,scaleFactor=1.1, minNeighbors=3)
      color = (255, 255, 255) #検出範囲を囲う線の色を指定
      print("検出対象ファイル："+file_name)
      print("検出された顔の数："+str(len(face)))
      #検出された顔カウンタ
      i=1
      #検出範囲の枠を書き込んだ画像の保存先パス
      surrounded_image_path =surrounded_dir +'/'+root+'_surrounded' + ext
      if len(face) > 0:  
        for rect in face:
          x=rect[0]
          y=rect[1]
          width=rect[2]
          height=rect[3]
          dst=img[y:y+height,x:x+width]
          #検出画像サイズを200×200にリサイズ（不要な場合はコメントアウトしてください）
          dst=cv2.resize(dst,(200,200))
          #検出された顔画像の保存
          crop_image_path=crop_dir+'/'+root+'_'+str(i)+ext
          cv2.imwrite(crop_image_path,dst)
          i+=1
          #元画像に検出範囲の枠を書き込む
          cv2.rectangle(img,tuple(rect[0:2]),tuple(rect[0:2]+rect[2:4]),color,thickness=2)
        cv2.imwrite(surrounded_image_path,img)
      else:
        print("no face")
        n+=1
        n_files.append(file_name)
        cv2.imwrite(surrounded_image_path,img)
               
```

```
print("顔が検出されなかった画像数："+str(n))
print("顔が検出されなかった画像ファイル名：",n_files)
```

```
#顔が検出されなかった画像ファイル名の書き出し
if os.path.exists(surrounded_dir+'/n_files.txt'):
    print("書き出し先ファイルに追記します。")
    with open(surrounded_dir+'/n_files.txt','a') as f:
      f.write("--検出されなかった画像一覧--\n")
    for n_file in n_files:
      with open(surrounded_dir+'/n_files.txt','a') as f:
        f.write(n_file+'\n')
else:
    print("書き出し先ファイルがないので新規作成します")
    with open(surrounded_dir+'/n_files.txt','w') as f:
      f.write("--検出されなかった画像一覧--\n")
    for n_file in n_files:
      with open(surrounded_dir+'/n_files.txt','a') as f:
        f.write(n_file+'\n') 
```
